```python
#make a dataframe with algorithms as rows and datasets as cols and ordered by a specific ordering over molecules
# get 1781 values for merged
```

```python
import pandas as pd
from scipy.stats import wilcoxon
```

```python
algorithms = ['elastic','ksvm','ridge','rf']
datasets = ['f','p','m']

columns = []
for a in algorithms:
    for d in datasets:
        columns.append(a+'_'+d)

f = open('rf_f','r')
labels = []
for line in f:
    line = line.strip()
    mol,val = line.split(":")
    labels.append(mol[6:])


df = pd.DataFrame(columns=columns,index = labels)
```

```python
for algo in algorithms:
    
    for data in datasets:
        
        f = open(algo+"_"+data,'r')
        
        for line in f:
            
            line = line.strip()
            mol,val = line.split(":")
            mol = mol[6:]
            
            col = algo+'_'+data
            
            df.loc[mol,col] = val
```

```python
# Do not drop once you get merged values for 1781
df.drop(labels='1781',axis=0,inplace=True)
```

```python
wins_summary = pd.DataFrame(columns = columns, index = columns)

# number of times row index wins against column index in total of 50 attempts ( currently 49)
for c1 in columns:
    for c2 in columns:
        
        row_data = df[c1]
        columns_data = df[c2]
        
        wins = [row_data[i] < columns_data[i] for i in range(len(row_data))]
        wins_summary.loc[c1,c2] = sum(wins)
```

```python
wins_summary
```

```python
wilcoxon_p = pd.DataFrame(columns = columns, index = columns)
wilcoxon_t = pd.DataFrame(columns = columns, index = columns)

for c1 in columns:
    for c2 in columns:
        
        if(c1==c2):
            continue
        
        row_data = pd.to_numeric(df[c1])
        columns_data = pd.to_numeric(df[c2])
        
        wilcoxon_p.loc[c1,c2] = wilcoxon(row_data,columns_data)[1]>0.05
        wilcoxon_t.loc[c1,c2] = wilcoxon(row_data,columns_data)[0]
```
